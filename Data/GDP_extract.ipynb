```python
def decimal_to_quarter(val):
    year = int(val)
    fraction = val - year

    quarter = int((fraction + 0.01) * 4) + 1
    
    return f"{year}Q{quarter}"

# 3. Apply it to create a new column
df_subset['date'] = df_subset['timeQtr'].apply(decimal_to_quarter)


print(df_subset)
```
